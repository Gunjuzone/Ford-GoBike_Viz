## Investigation Overview
> The dataset was analysis to get some insights. It was discovered that, there are two types of bike users; subscibers and custoners. Customer  spend more time cycling than subscribers but subscribers are more consistent with cycling. 

> - The information provided on the gnder of the users showed that, male genders spend more time cycling than female and other genders.

## Dataset Overview
> - This data set includes information about individual rides made in Ford Go Bike system covering the greater San Francisco Bay area, United States. It contains columns like the type of users, gender of the users, duration of ride and age of the users.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('fordgobike_clean.csv')
```

## BIKE DAILY USAGE

> Do you know, in San Francisco Bay Area, more people ride bike on thursday than any other day?

```python
# Generate day of teh week in order
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
#Bike Daily Usage
count = df['start_day'].value_counts()
order = count.index
base_color = sb.color_palette()[0]
plt.figure(figsize=(10,6))
sb.countplot(data=df, x='start_day',color=base_color, order = day_order);
plt.ylabel('Trip Counts',fontsize = 14)
plt.xlabel('Days of the week',fontsize = 14)
plt.xticks(rotation=90)
plt.title('Ford Go Bikes System Daily Usage',fontsize = 22);
```

## AVERAGE CYCLING TIME

> While thursday is the busiest day of the week, people spend more time cycling on weekends than weekdays, including thursday.

```python
#Check average minutes of cycling per day
plt.figure(figsize = (10, 6))
sb.pointplot(data = df, x = 'start_day', y = 'duration_min', order = day_order)
plt.xticks(rotation = 90)
plt.title('Average Daily Ride', fontsize = 25)
plt.ylabel('Average Duration (min)', fontsize = 14)
plt.xlabel('Days', fontsize = 14);
```

## Gender of the most bike user


> Whilst, customer spend longer time riding than subscriber, neither male or female gender cycle the most, see the next slide for who cycle the most on average.

```python
#PLot distance cycle by gender and user type
plt.figure(figsize = (10, 7))
sb.barplot(data=df, x='user_type', y='duration_min', hue='member_gender', palette='colorblind')
plt.xlabel('User type', fontsize= 16)
plt.ylabel('Average duration (min)', fontsize= 16)
plt.title('Duration for cycling for different gender of bike users', fontsize= 20);
```
